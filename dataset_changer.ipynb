```python
data = []
cnt = 0
sample_cnt = -1
label = 0
with open('rocket_league_skillshots.data') as f:
  features = next(f).split()
  temp_data = []
  d = dict()
  for id_line, line in enumerate(f):
    split_line = line.split()
    if (len(split_line) == 1):
      sample_cnt += 1
      label = int(split_line[0])
      d[sample_cnt] = {'label': label, 'time_series':[]}
      continue
    elif (len(split_line) != len(features)):
      print("This line doesn't match the expected number of features!")
    else:
      temp_dict = {}
      for f, val in enumerate(split_line):
        if f < 7:
          temp_dict[features[f]] = float(val)
        else:
            temp_dict[features[f]] = int(val)
      d[sample_cnt]['time_series'].append(temp_dict)
```
